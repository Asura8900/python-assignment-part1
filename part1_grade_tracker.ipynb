***FOR TASK 1***: Given raw student data 

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

*Requirement 1*: To Loop through raw_students to produce a cleaned version for each student

```python
students_clean_data = [] # Creating a empty list
i = 0

for student in raw_students:
    students_clean_data.append({}) # To add a dictionary inside empty list
    clean_name = student["name"].strip().title() # leading/trailing whitespace removed and is converted to Title Case
    students_clean_data[i].setdefault("name", clean_name) # Adding to dictionary inside list
    roll_int = int(student["roll"]) # converted from a string to an integer
    students_clean_data[i].setdefault("roll", roll_int) # Adding to dictionary inside list
    marks_cart = [] # Creating an empty list to add marks in int-type
    for mark_str in student["marks_str"].split(", "):  #marks_str is split on ", " and each element is converted to an integer, producing a marks_cart list
        mark_int = (int(mark_str)) # convert to int-type
        marks_cart.append(mark_int) # adding to mark_cart list
    students_clean_data[i].setdefault("marks", marks_cart) # Adding to dictionary inside list
    i += 1
```

*Requirement 2*: To Verify the name contains only alphabetic characters. Print "✓ Valid name" or "✗ Invalid name" next to each student.

```python
# Making loop to validate name
for student in students_clean_data:  
    no_space_name = student["name"].replace(" ", "") # to remove spaces 
    if  no_space_name.isalpha() == True : # Verifying the name contains only alphabetic characters
        print(f'{student["name"]:15}  → ✓ valid name')
    else:
        print(f'{student["name"]:15} → ✗ invalid name')
```

*Requirement 3*: To print a formatted profile card 

```python

for student in students_clean_data:
    print("==================================")
    print(f'Student : {student["name"]}')
    print(f'Roll no : {student["roll"]}')
    print(f'Marks   : {student["marks"]}')
    print("==================================")
   
```

*Requirement 4*: To print the name of roll number 103 in ALL CAPS and lowercase.

```python

for student in students_clean_data:
    if student["roll"] == 103:
        print(student["name"].upper())
        print(student["name"].lower())
        
```

***FOR TASK 2***: Given data

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

*Requirement 1*: To print each subject alongside its marks and a grade label

```python

print("=================================")
for i in range(0,len(subjects)):
    mark = marks[i]
    if 90 <= mark <=100 :
        print(f'{subjects[i]:10}: {marks[i]:5}  →  Grade A+')
    elif 80 <= mark < 90 :
        print(f'{subjects[i]:10}: {marks[i]:5}  →  Grade A')
    elif 70 <= mark < 80 :
        print(f'{subjects[i]:10}: {marks[i]:5}  →  Grade B')
    elif 60 <= mark < 70 :
        print(f'{subjects[i]:10}: {marks[i]:5}  →  Grade C')
    else:
        print(f'{subjects[i]:10}: {marks[i]:5}  →  Grade F')
print("=================================")
```

*Requirement 2*: To calculate total marks, Average marks, Highest scoring subject, Lowest scoring subject

```python

print("=======================================")
print(f'Total marks             : {sum(marks)}') # Total marks (sum of all subjects)
print(f'Average marks           : {round(sum(marks)/ len(marks),2)}') # Average marks (rounded to 2 decimal places)
print(f'Highest scoring subject : {subjects[marks.index(max(marks))]} ({max(marks)})') # Highest scoring subject (name + marks)
print(f'Lowest scoring subject  : {subjects[marks.index(min(marks))]} ({min(marks)})') # Lowest scoring subject (name + marks)
print("=======================================")
```

*Requirement 3*: Marks-entry system

```python
# Creating an Exception error
class OutOfRange(Exception):
    pass

new_sub = 0
# making an endless loop unless told "done" to stop
while True:
    subject = input("Enter subject name (or type 'done'): ")
    if subject.lower() == 'done':
        break
    mark = input("Enter marks (0-100)")
    try:
        mark_int = int(mark)
        if 0 <= mark_int <= 100:
            subjects.append(subject)
            marks.append(mark_int)
            new_sub += 1
        else :
            raise OutOfRange("do not add invalid entries to the list.")
    except OutOfRange as e:
        print(e)
    except ValueError:
        print("do not crash")

print("=====================================")
print(f'New subjects added: {new_sub}')
print(f'Updated average marks: {round(sum(marks)/ len(marks),2)}')
print("=====================================")
```

***FOR TASK 3***: Given class data 

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

*Requirement 1 & 2*: To print a formatted class report

```python
# Creating variables to store the values for further use
pass_count = 0
fail_count = 0
topper_avg = 0
student_avg_sum = 0
# Conputing to assign value to variables & also print report
for name, marks in class_data:
    average = round(sum(marks)/len(marks),2)
    student_avg_sum += average
    if average >= 60:
        result_status = "Pass"
        pass_count += 1
    else:
        result_status = "Fail"
        fail_count += 1
    if average > topper_avg:
        topper_avg = average
        topper_name = name
    print(f'{name:20} | {average:6.2f} | {result_status}')
    
```

*Requirement 3*: To calculate number of students who passed and who failed, Class topper and Class average 

```python
# using stored variable 
print("=====================================")
print(f'Passed : {pass_count}    |    Failed: {fail_count}') 
print(f'Topper       : {topper_name} {topper_avg:.2f}')
print(f'Class average: {round(student_avg_sum/len(class_data),2)}')
print("=====================================")
```

```python

clean_essay_split = clean_essay.split(". ")
print(clean_essay_split)
```
